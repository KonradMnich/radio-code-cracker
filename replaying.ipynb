```python
import pandas as pd
```

```python
df = pd.read_csv("szlaban u rodzicow 2022-08-10_v2.csv", index_col="time", parse_dates=True).dropna()
df
```

```python
code = df.bin
zeros = 0
ones = 0
last_digit = 0
compressed_array = []
for i, x in enumerate(code):
    if x == 0 and last_digit == 0:
        zeros += 1
    elif x == 1 and last_digit == 0:
        compressed_array.append({"digit": 0, "occurences": zeros, "timestamp": code.index[i]})
        zeros = 0
    elif x == 1 and last_digit == 1:
        ones += 1
    elif x == 0 and last_digit == 1:
        compressed_array.append({"digit": 1, "occurences": ones, "timestamp": code.index[i]})
        ones = 0
    last_digit = x
        
df2 = pd.DataFrame.from_records(compressed_array)
df2
```

```python
df2["duration"] = df2.timestamp.diff()
df2
```
